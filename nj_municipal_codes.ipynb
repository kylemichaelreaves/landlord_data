```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/kylereaves/Desktop/nj_municipal_codes.csv', header=None)
```

```python
df['code'] = df[0].str.extract(r'(\d+)')
df['municipality'] = df[0].str.extract(r'([A-Z]+\s+[A-Z]+)')
```

```python
df.drop(columns=[0], inplace=True)
```

```python
df['municipality'] = df.municipality.str.title()
```

```python
df.index = df['code']
```

```python
df.drop(columns=['code'], inplace=True)
```

```python
df.to_csv('')
```
